```python
import matplotlib.pyplot as plt
from matplotlib.text import Text
import numpy as np 
import scienceplots
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
plt.style.use('science')
```

```python
hpica = 14 
wpica = 42
hinch = hpica/6
winch = wpica/6

fig = plt.figure(figsize=(winch, hinch), dpi=300)
subfigs = fig.subfigures(2,1).flatten()

ax0 = subfigs[0].subplots()
label_a = Text(0.01, 0.99, 'A', fontsize=12, transform=subfigs[0].transSubfigure, ha="left", va="top")
subfigs[0].add_artist(label_a)

ax1 = subfigs[1].subplots()
label_b = Text(0.01, 0.99, 'B', fontsize=12, transform=subfigs[1].transSubfigure, ha="left", va="top")
subfigs[1].add_artist(label_b)

fig.savefig('gl.eps')
```
